```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('companies.db')
```

```python
companies_df = pd.read_csv('companies.csv', sep=';')
companies_df.head()
```

```python
companies_df.to_sql(name='companies', con=con)
```

```python
cursor = con.cursor()
```

```python
select_all = "SELECT * FROM companies"
rows = cursor.execute(select_all).fetchall()
rows
```
